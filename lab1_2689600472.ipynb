```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)

# จำนวนข้อมูล
n = 100

# สร้าง dataset
data = pd.DataFrame({
    "student_id": range(1, n+1),
    "sleep_hours": np.random.normal(7, 1.5, n).clip(3, 10),
    "study_hours": np.random.normal(4, 2, n).clip(0, 10),
    "screen_time": np.random.normal(5, 2, n).clip(1, 12),
    "exercise_minutes": np.random.normal(30, 15, n).clip(0, 90),
})

# mood score (ให้สัมพันธ์กับการนอน)
data["mood_score"] = (
    data["sleep_hours"] * 0.8
    - data["screen_time"] * 0.2
    + np.random.normal(3, 1, n)
).clip(1, 10)

# save csv
data.to_csv("student_sleep.csv", index=False)
print(data.head())
```

```python
sns.set(style="whitegrid")

# Histogram sleep
plt.figure(figsize=(6,4))
plt.hist(data["sleep_hours"], bins=15)
plt.title("Sleep Hours Distribution")
plt.xlabel("Sleep Hours")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x="sleep_hours", y="mood_score", data=data)
plt.title("Sleep Hours vs Mood Score")
plt.show()
```
